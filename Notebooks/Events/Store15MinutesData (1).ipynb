```
configs = {"fs.azure.account.auth.type": "OAuth",
          "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
          "fs.azure.account.oauth2.client.id": app_reg_client_id,
          "fs.azure.account.oauth2.client.secret": app_reg_secret_value,
          "fs.azure.account.oauth2.client.endpoint": app_reg_endpoint}

try:
    dbutils.fs.mount(
    source = source_path,
    mount_point = mount_point,
    extra_configs = configs)
except:
    print('Directory already mounted')
```

```
import os
if os.path.exists('/dbfs/'+landing):
    dbutils.fs.rm(landing,recurse=True)
```

```
from pyspark.sql.functions import *
df=spark.read.format('delta').load(mount_point+"/decodeddf")
```

```
max_timestamp=df.select(max('TimeStamp')).rdd.map(lambda x:x[0]).collect()[0]
print(max_timestamp)
```

```
df.orderBy('Time').filter(col('TimeStamp')> max_timestamp-expr(duration)).coalesce(1).write.format('csv').option('header','true').mode('overwrite').save(landing)
```

```
file=dbutils.fs.ls(landing)
for i in file:
    if i.name.startswith('part'):
        dbutils.fs.cp(landing+i.name,landing+'/1HourLatest.csv',recurse=False)
    dbutils.fs.rm(landing+i.name)
print('DATA WRITTEN')        
```
